```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('vaccination-data.csv')
```

```python
df.head()
```

```python
df['COUNTRY'].value_counts()
```

```python
df.isna().sum()
```

```python
df["FIRST_VACCINE_DATE"].fillna("2021-01-01", inplace = True)
```

```python
df.isna().sum()
```

```python
plt.figure(figsize=(15, 40))
df['VACCINES_USED'].value_counts().head(25).plot(kind='pie')
plt.show()
```

```python
df = df.dropna(axis = 0)
```

```python
df
```

```python
df.isna().sum()
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'ISO3',
            y = 'TOTAL_VACCINATIONS',
            data = df.head(25))
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'ISO3',
            y = 'TOTAL_VACCINATIONS_PER100',
            data = df.head(25))
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'ISO3',
            y = 'PERSONS_VACCINATED_1PLUS_DOSE_PER100',
            data = df.head(25))
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'ISO3',
            y = 'PERSONS_FULLY_VACCINATED_PER100',
            data = df.head(25))
```
